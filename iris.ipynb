```python
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
from sklearn.metrics import accuracy_score 
%matplotlib inline
```

```python
cloums=["sepal_length","sepal_width","petal_length","petal_width","Class_lables"]
df = pd.read_csv('/content/sample_data/IRIS.csv',names=cloums)#
print(df['Class_lables'])
```

```python
sns.pairplot(df, hue='Class_lables')
```

```python
data = df.values
x=data[:,0:4]
y=data[:,4:]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
pr=model.predict(x_test)
print(accuracy_score(y_test,pr)*100)
```

```python
x_new =[[5,2,3,1]]
pr=model.predict(x_new)
print(pr)
```
